```python
def get_ramsey_contrast(chain, ion_num, times, creation_op_num = 0, time_precision=1.e-6):
        """Return an array of amplitudes of ion ion_num at each time in time_interval   

        """
        eigen = LA.eig(chain.get_couplings())
        eigenvectorsT = eigen[1]
        eigenfreqs   = eigen[0]

        def gamma(l, j, t, chain):
            
            s = 0
            for k in range(chain.num_of_ions):
                s += eigenvectorsT[:,k][l] * eigenvectorsT[:,k][j] * np.exp(1.j* eigenfreqs[k] * t)
            return s
        
        #time_interval = [0., 700.e-6]
        #times = np.arange(0., 700.e-6, time_precision)

        
        l = ion_num
        j = creation_op_num
        ex = abs(np.conjugate(np.array([gamma(l, j, t, chain) for t in times])) * np.array([gamma(l, j, t, chain) for t in times]))
        return ex, np.array([gamma(l, j, t, chain) for t in times])
```
